USING A PERCEPTRON ON THE CLASSIC IRIS DATASET TO IDENTIFY A PIRTICULAR SPECIES OF FLOWER

```python
iris = load_iris(as_frame=True)
iris
```

```python
import matplotlib.pyplot as plt
iris_df = sklearn.datasets.load_iris()
iris_df
```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()
scatter = ax.scatter(iris_df.data[:, 0], iris_df.data[:, 1], c=iris_df.target)
ax.set(xlabel=iris_df.feature_names[0], ylabel=iris_df.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris_df.target_names, loc="lower right", title="Classes"
)
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(iris_df.data[:, 2], iris_df.data[:, 3], c=iris_df.target)
ax.set(xlabel=iris_df.feature_names[2], ylabel=iris_df.feature_names[3])
_ = ax.legend(
    scatter.legend_elements()[0], iris_df.target_names, loc="lower right", title="Classes"
)
```

SINCE THE PETAL LENGTH ARE ALMOST LINERLY SAPERABLE WE USE THESE FEATURES AS INPUTS TO THE PRECEPTRON MODEL

```python
X = iris.data[['petal length (cm)','petal width (cm)']].values
y = (iris.target == 0)
```

```python
y
```

```python
iris.target
```

```python
per_1 = Perceptron(random_state=42)
per_1.fit(X,y)
```

```python
X_test = [ [2,0.5] , [3,1] ,[4,1.5] ]
y_preds = per_1.predict(X_test)
y_preds
```
